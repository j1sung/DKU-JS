## 1. 기본 그래프 도구
* 선 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('시도별 전출입 인구수.xlsx')
```

```python
df.head()
```

```python
# 누락값을 앞 데이터로 채움
df = df.fillna(method = 'ffill')
```

```python
df.head()
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]  # 데이터필터링
df_seoul = df_seoul.drop(['전출지별'], axis = 1)
df_seoul.rename({'전입지별' : '전입지'}, axis = 1, inplace = True)
df_seoul.set_index('전입지', inplace = True)
```

```python
df_seoul
```

```python
df_seoul
```

```python
# 서울 -> 경기도로 이동한 인구 데이터 값만 출력
sr_one = df_seoul.loc['경기도']
```

```python
sr_one.head()
```

```python
from matplotlib import font_manager, rc
font_path = "./malgun.ttf"
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)

sr_one = df_seoul.loc['경기도']
plt.plot(sr_one.index, sr_one.values)  # x,y축 데이터를 plot함수에 입력
plt.plot(sr_one)
```

```python
sr_one = df_seoul.loc['경기도']
plt.plot(sr_one.index, sr_one.values)

plt.title('서울 -> 경기도 인구 이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.show()
```

```python
sr_one = df_seoul.loc['경기도']
plt.figure(figsize = (14, 5))  # 그림 사이즈 지정
plt.xticks(rotation = 'vertical')  # x축 눈금 라벨 회전하기

plt.plot(sr_one.index, sr_one.values)

plt.title('서울 -> 경기도 인구 이동')
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.legend(labels = ['서울 -> 경기'], loc = 'best')

plt.show()
```

```python
sr_one = df_seoul.loc['경기도']
plt.style.use('ggplot')
plt.figure(figsize = (14, 5))
plt.xticks(size = 10, rotation = 'vertical')

plt.plot(sr_one.index, sr_one.values, marker = 'o', markersize = 10)

plt.title('서울 -> 경기도 인구 이동', size = 30)
plt.xlabel('기간', size = 20)
plt.ylabel('이동 인구수', size = 20)

plt.legend(labels = ['서울 -> 경기'], loc = 'best', fontsize = 15)

plt.show()
```
